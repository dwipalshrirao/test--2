# What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook

Correlation describes the relationship between two variables. 
It indicates how changes in one variable are associated with changes in another variable. 
Correlation can range between -1 to 1
where:
if correlation is +ve then we say that 2 veriables are positively correlated i.e. as one variable increases, the other variable also increases.
if correlation is 0 then we say that there is no relation between 2 veriables i.e. increase/decrease of one variable does not make change in the other variable.
if correlation is -ve then we say that 2 veriables are negatively correlated i.e. as one variable increases, the other variable decreases.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

x = [2, 4, 5, 6, 6, 7, 8]
y = [4, 16, 26, 37, 36, 49, 63]

data = pd.DataFrame({'x': x, 'y': y})

# Calculate correlation coefficient
correlation_coefficient = data['x'].corr(data['y'])
print("from given example we can say that 2 veriables are highly  correlated")
# Plot the data
sns.scatterplot(x='x', y='y', data=data)
plt.title(f'Correlation Coefficient: {correlation_coefficient:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a sample dataset with no relationship
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Calculate correlation coefficient
correlation_coefficient = data['X'].corr(data['Y'])

print("from given example we can say that 2 veriables are not much related")
# Plot the data
sns.scatterplot(x='X', y='Y', data=data)
plt.title(f'Correlation Coefficient: {correlation_coefficient:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
