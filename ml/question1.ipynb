```python
import pandas as pd
```

```python
df = pd.read_csv("instagram_reach.csv")
df
```

```python
def remove_special_chars(string):
    return ''.join(letter for letter in string if letter.isalnum() or letter == " ")
```

```python
# df["Caption"].str.split(" ")
df["Caption"] =df["Caption"].fillna("")
df["Caption_splited"] = df["Caption"].apply(remove_special_chars).str.split(" ")
```

```python
# df["Hashtags"].str.split("#")
df["Hashtags"] =df["Hashtags"].fillna("")
df["Hashtags_splited"] = df["Hashtags"].apply(remove_special_chars).str.split(" ")
```

```python
df
```

```python
df["string_feature"] = df.apply(lambda x: x["Caption_splited"] + x["Hashtags_splited"] , axis=1) 
df["string_feature"]
```

```python
df = df.explode('string_feature')
```
